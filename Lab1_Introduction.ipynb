# Sequences
### Strings, Tuple and List
* A **string** is a sequence enclosed within single, double or tripple qoutation. Remember, string is **immutable**, which means if you perform an operation on strings, you always produce a new string.
* **Tuple** is a sequence of comma separated items, enclosed in parentheses ().
* **List** is also a sequence of comma seperated items, enclosed in parenthesis [ ].

The main difference between tuple and list is that list is mutable while tuple is not. `len()` function gives the total number of items in a string, tuple or list.



````python
Strng = 'MAT 120'                                    # A string
Tpl   = ('MAT', '120', 120, 20)                      # It's a tuple
Lst   = [333, 'Python', 'is', 'slow']                # It's a List

print(Tpl[0], Lst[1])
print(Lst[1:4])

Lst.append(666)                                      # It inserts the number 666 at the end of the list
print(Lst)

Lst.insert(3, 'not')                                 # It inserts the 'not' at some chosen position of the list
print(Lst)

len(Strng), len(Tpl), len(Lst)
````

###Range function: ```range(start, start, step)```. `range()` is a built-in function used to generate a sequence of numbers.
- start: Optional. The starting value of the sequence (default is 0).
- stop: Required. The end value of the sequence (not inclusive).
- step: Optional. The step value between each number in the sequence (default is 1). \\

`range()` only take integers in its arguments.

````python
for i in range(0, 5):
  print(i)
````

````python
for i in range(0, 5, 2):
  print(i)
````

#Array
Use library `numpy` to create array from a list. Unlike lists, all elements of an array must have the same data-type. Arrays are 50 times faster than Lists.

````python
import numpy as np

list1, list2 = [2, -3, 4], [3, 1, 1]
arr1 = np.array(list1)
arr2 = np.array(list2)

dot  = np.dot(arr1, arr2)                           # This is the dot product
print('Dot Product:', dot)
````

## Special arrays
`numpy` has `arange()` function which works like the `range()` function but can take float in its argument. \\
- `numpy.arange(start, stop, step)`

We can use `numpy.linspace()` to create the same array instead. The only difference is that it takes **Number of Points** in the third argument rather than the **increments**.
- `numpy.linspace(start, stop, number)`

It's important to note that
$$\text{step}= \frac{\text{stop} - \text{start}}{\text{number} -1} $$


````python
np.arange(0, 10, 0.1)
````

````python
np.linspace(0, 10, 101)
````

#Plotting
Use `matplotlib` library to plot graphs.

````python
import matplotlib.pyplot as plt

def f(x): return np.sin(x)

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

plt.plot(x, f(x))                                 # Plots the graph
plt.title("x vs. sin(x) curve")                   # It's just a title which is a string. You don't need $ symbols here.
plt.xlabel("x")                                   # Labels the x axis
plt.ylabel("sin(x)")                              # labels the y axis
plt.axhline(color = 'black')                      # Creates x axis as a horizontal line
plt.axvline(color = 'black')                      # Creates y axis as a vertical line
plt.grid()
plt.show()                                        # It shows the plot.
````
